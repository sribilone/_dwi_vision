```python
# Import libraries
import pandas as pd
import os
import shutil
```

# Original Image

```python
original_path = '/Users/Sribilone/AiiLAB/_dwi_project/original_dwi_image'
```

```python
for set_folder in os.listdir(original_path):
    # Skip folder when set_folder name in list.
    if set_folder in ['.DS_Store']:
        print(f"Skip reading '{set_folder}'")
        continue

    if set_folder == 'SET_1':
        destination = '/Users/Sribilone/AiiLAB/_dwi_project/image_original_dataset/positive_case'
    elif set_folder == 'SET_2':
        destination = '/Users/Sribilone/AiiLAB/_dwi_project/image_original_dataset/negative_case'

    
    # Setting path
    set_path = os.path.join(original_path, set_folder)
    if os.path.isdir(set_path):
        print(f"Reading folder: '{set_folder}' >> {set_path}")
        for img_folder in os.listdir(set_path):
            if img_folder in ['.DS_Store']:
                print(f"*>> Skip reading {set_folder} ==> '{img_folder}'")
                continue
            print(f"{set_folder} >> {img_folder}")

            # Setting path
            img_path = os.path.join(set_path, img_folder)
            if os.path.isdir(img_path):
                for img_file in os.listdir(img_path):
                    if img_file in ['.DS_Store']:
                        print(f"*>> Skip reading {img_folder} ==> '{img_file}'")
                        continue
                    img_file_path = os.path.join(img_path, img_file)
                    shutil.copy(img_file_path, destination)
                    print(f"Copy from {set_folder}/{img_folder}/{img_file} to {destination}")
                    new_name = f"{img_folder}_{img_file}"
                    copied_file = os.path.join(destination, img_file)
                    renamed_file = os.path.join(destination, new_name)
                    os.rename(copied_file, renamed_file)
```
